```python
elements = [10, 7, 26, 43, 3, 13, 11]
```

```python
#Bubble sort 
def b_sort(elements):
    for i in range(len(elements)-1,0,-1):
        for j in range(i):
            if (elements[j]>elements[j+1]):
                temp=elements[j]
                elements[j]=elements[j+1]
                elements[j+1]=temp
    
b_sort(elements)
print("Bubble Sort:",elements)
```

```python
#Insertion Sort
def Insertion_sort(elements):
    for i in range(1, len(elements)):
        j = i
        while elements[j - 1] > elements[j] and j > 0:
            elements[j - 1], elements[j] = elements[j], elements[j - 1]
            j -= 1

Insertion_sort(elements)
print("InsertionSort:",elements)
```

```python
#•	Write a python program that will search for a specific element using binary search.
n = 10
pos = -1 
def binary_search(elements, n):
    lowElement = 0
    upperElement = len(elements) - 1
    while lowElement <= upperElement:
        middleElement = (lowElement + upperElement) // 2
        if list[middleElement] == n:
            globals()['pos'] = middleElement
            return True
        else:
            if elements[middleElement] < n:
                lowElement = middleElement + 1
            else: 
                upperElement = middleElement  - 1
    return False

if binary_search(elements, n):
    print("Found", pos + 1)
else:
    print("Not")
```
